```python
import matplotlib 
import matplotlib.pyplot as plt
import plotly.express as px
import pandas as pd
import geopandas as gpd
import numpy as np
import folium
```

Karten für die Welt und Europa und interaktive Karten für deutsche Städte und Thüringen Städte.

```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
```

```python
world.plot()
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(20,10)
plt.title('Weltkarte')
```

Europa Koordinaten

```python
Europa = world.loc[world['continent'] == 'Europe']
Europa
```

```python
Europa.plot()
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(20,10)
plt.title('Europa')
```

Deutsche Städte

```python
karte_path1 = r'C:\Users\Dell\Desktop\Geospatial Analysis Projekte\Geodaten\de.csv'
df1 = pd.read_csv(karte_path1)
df1
```

```python
Deutschland_Städte = df1.loc[:, ("city", "lat", "lng")]
Deutschland_Städte
```

Koordinaten

```python
Deutschland_Städte_Koordinaten = gpd.GeoDataFrame(Deutschland_Städte, geometry = gpd.points_from_xy(Deutschland_Städte.lat, Deutschland_Städte.lng))
Deutschland_Städte_Koordinaten['geometry']
```

Thüringen Städte

```python
karte_path2 = r'C:\Users\Dell\Desktop\Geospatial Analysis Projekte\Geodaten\Thuringen-data-table-export.csv'
df2 = pd.read_csv(karte_path2)
df2
```

```python
Thüringen_Städte = df2.loc[:, ("Town Name", "Latitude", "Longitude")]
Thüringen_Städte
```

Koordinaten

```python
Thüringen_Städte_Koordinaten = gpd.GeoDataFrame(Thüringen_Städte, geometry = gpd.points_from_xy(Thüringen_Städte.Latitude, Thüringen_Städte.Longitude))
Thüringen_Städte_Koordinaten['geometry']
```

interaktive Karten

```python
Deutschland = [51.165691, 10.451526]
Deutschland_karte = folium.Map(Deutschland, zoom_start = 15, tiles = 'Stamen Terrain')
Deutschland_karte
```

```python
for i in range (0, len(Deutschland_Städte)):
    folium.Marker([Deutschland_Städte.iloc[i]['lat'], Deutschland_Städte.iloc[i]['lng']],
                    popup = Deutschland_Städte.iloc[i]['city']).add_to(Deutschland_karte)
Deutschland_karte
```

```python
Deutschland_Thüringen = [51.207211, 10.454490]
Deutschland_Thüringen = folium.Map(Deutschland_Thüringen, zoom_start = 15, tiles = 'Stamen Terrain')
Deutschland_Thüringen
```

```python
for i in range (0, len(Thüringen_Städte)):
    folium.Marker([Thüringen_Städte.iloc[i]['Latitude'], Thüringen_Städte.iloc[i]['Longitude']],
                    popup = Thüringen_Städte.iloc[i]['Town Name']).add_to(Deutschland_Thüringen)
Deutschland_Thüringen
```
